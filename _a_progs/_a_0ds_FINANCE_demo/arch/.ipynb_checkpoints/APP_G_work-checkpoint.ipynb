```python
## CHECK DIV WORK

# CHECK with total DIVIDENDS
import yfinance as yf
import pandas as pd

# Assuming df_div is preloaded and formatted correctly

# Initialize a cache for dividends to minimize API calls
dividend_cache = {}

# Function to fetch and cache dividends
def get_cached_dividends(stock):
    if stock not in dividend_cache:
        ticker = yf.Ticker(stock)
        dividends = ticker.dividends
        dividends.index = dividends.index.tz_localize(None)  # Ensure timezone-naive
        dividend_cache[stock] = dividends
    return dividend_cache[stock]

# Adjusted function to use cached dividends
def get_dividends(stock, start_date, end_date):
    dividends = get_cached_dividends(stock)
    relevant_dividends = dividends[(dividends.index >= start_date) & (dividends.index <= end_date)]
    return relevant_dividends.sum()

# Calculate total dividends with potential optimizations
df_div['total_dividends'] = df_div.apply(lambda row: get_dividends(row['Stock'], row['StartDate'], row['EndDate']) * row['units_hold'], axis=1)
df_div = df_div.sort_values(by=['Stock', 'StartDate', 'EndDate'])

print(df_div)

df_div['total_dividends'].sum()
```

```python
# CHECK with total DIVIDENDS
import yfinance as yf
import pandas as pd

# Assuming df_div is preloaded and formatted correctly

# Initialize a cache for dividends to minimize API calls
dividend_cache = {}

# Function to fetch and cache dividends
def get_cached_dividends(stock):
    if stock not in dividend_cache:
        ticker = yf.Ticker(stock)
        dividends = ticker.dividends
        dividends.index = dividends.index.tz_localize(None)  # Ensure timezone-naive
        dividend_cache[stock] = dividends
    return dividend_cache[stock]

# Adjusted function to use cached dividends
def get_dividends(stock, start_date, end_date):
    dividends = get_cached_dividends(stock)
    relevant_dividends = dividends[(dividends.index >= start_date) & (dividends.index <= end_date)]
    return relevant_dividends.sum()

# Calculate total dividends with potential optimizations
df_div['total_dividends'] = df_div.apply(lambda row: get_dividends(row['Stock'], row['StartDate'], row['EndDate']) * row['units_hold'], axis=1)
df_div = df_div.sort_values(by=['Stock', 'StartDate', 'EndDate'])

print(df_div)
```

```python
from datetime import datetime, timedelta

# Function to calculate periods' end dates (inclusive of today)
def calculate_period_dates():
    today = datetime.now()
    # Ensure we're including the entire day for "today" by setting the time to the end of the day
    today_end = today.replace(hour=23, minute=59, second=59)
    three_months_ago = today - timedelta(days=90)
    six_months_ago = today - timedelta(days=180)
    twelve_months_ago = today - timedelta(days=365)
    return today_end, three_months_ago, six_months_ago, twelve_months_ago

# Adjust the aggregate_dividends function to ensure correct period handling
def aggregate_dividends(df, period_dates):
    today_end = period_dates[0]
    aggregated_dividends = []
    for date in period_dates[1:]:  # Skip the 'today_end' for the loop, but use it as the end date
        total_dividends = 0
        for index, row in df.iterrows():
            start_date = date.strftime("%Y-%m-%d")
            # Use the end of today for all calculations to include dividends up until today
            end_date = today_end.strftime("%Y-%m-%d")
            total_dividends += get_dividends(row['Stock'], start_date, end_date) * row['units_hold']
        aggregated_dividends.append(total_dividends)
    return aggregated_dividends

# Now, ensure the "print_dividends_report" function uses the updated logic correctly.
def print_dividends_report(df):
    period_dates = calculate_period_dates()
    # We'll now prepend the "Total dividends up until today" calculation
    total_dividends_today = sum(df.apply(lambda row: get_dividends(row['Stock'], datetime.now().strftime("%Y-%m-%d"), period_dates[0].strftime("%Y-%m-%d")) * row['units_hold'], axis=1))
    aggregated_dividends = [total_dividends_today] + aggregate_dividends(df, period_dates)
    report_dates = ["Up Until Today", "3 Months Ago", "6 Months Ago", "12 Months Ago"]
    for i, total_dividends in enumerate(aggregated_dividends):
        print(f"Total dividends {report_dates[i]}: {total_dividends}")

# Proceed to print the dividends report with the adjusted function
print_dividends_report(df_div)
```
