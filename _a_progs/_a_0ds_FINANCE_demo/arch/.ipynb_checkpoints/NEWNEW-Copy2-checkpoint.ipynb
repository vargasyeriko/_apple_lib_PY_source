```python
# UNITS
df_name = f'_df_0_init_UNITS.pkl'
df_add_name = '_df_add_1_up_to_date_UNITS.pkl'




import pandas as pd
import datetime
from datetime import datetime

### FNS
from _fns import sum_units_sold_by_stock_bought,sum_units_sold_by_stock_sold,sum_buy_sell_ordered

# Read cash Excess
data = {
    'date_cash_to_stock': ['2023-09-14', '2024-01-01'],
    'per_cash_to_stock': [30, 20],
    'stock_from_cash': ['BUGG.AX', 'BUGG.AX']
}
df_cash_x_stock = pd.DataFrame(data).sort_values(by='date_cash_to_stock').reset_index(drop=True)

# Initialize last cutoff date for filtering
last_cutoff_date = pd.Timestamp.min  # Use the smallest possible timestamp

# Loop through each row in df_cash_x_stock
for index, row in df_cash_x_stock.iterrows():
   

    # READ UNITS
    df = pd.read_pickle(f'_1_tables/{df_name}')
    # READ EXCHANGE TRANS HIST
    df_add = pd.read_pickle(f'_1_tables/{df_add_name }')
    df_add_original_lenght = len(df_add) # get # of trans before cutting df_add

    
    
    df_add['DateTrans'] = pd.to_datetime(df_add['DateTrans'])    
    
    date_cash_to_stock = pd.to_datetime(row['date_cash_to_stock'])
    
    
    
    # Filter df_add to include only new rows since the last processed date
    df_add = df_add[(df_add['DateTrans'] > last_cutoff_date) & (df_add['DateTrans'] <= date_cash_to_stock)]
    last_cutoff_date = date_cash_to_stock  # Update the last cutoff date
    last_cutoff_date_str = last_cutoff_date.strftime('%Y-%m-%d') #date arrangement


    # Optional: Save the filtered DataFrame if needed
    #filtered_df_add.to_pickle(f'_1_tables/_df_1_processed_{date_cash_to_stock.strftime("%Y%m%d")}.pkl')

    # Print updates
    print('\n|||>_updt.py_> > > Processing Portfolio UPDATES for up to :\n')
    print(f"Date Processed: {date_cash_to_stock.strftime('%Y-%m-%d')}, Rows processed: {len(df_add)}\n\n")
    df_cash_x_stock_ads = pd.DataFrame([row])
    print(pd.DataFrame([row]))  # Print the filtered DataFrame for this iteration

    # Placeholder for OPENING CASH EXCESS UPDTS
    filter_cash_excess_check = 1  # MEANS OPEN LOOP FOR UPDTS CASH EXCESS
    # #Correct the use of exec function
    with open("updt.py", "r", encoding="utf-8") as file:
        code = file.read()
        exec(code)
    #print(df)
    input('>>>>>>>>>>>>')

##### FOR EXTRA ROWS after last CASH EXCESS updt
    
##############################################################
print('\n\nCASH EXCESS ALGORITHM IS COMPLETELY PROCESSED',
      '\n\nenter to process rows after last CASH EXCESS\n\n')

df = pd.read_pickle(f'_1_tables/{df_name}')
# READ EXCHANGE TRANS HIST
df_add = pd.read_pickle(f'_1_tables/{df_add_name}')

df_add['DateTrans'] = pd.to_datetime(df_add['DateTrans'])    

# Filter df_add to include only new rows since the last processed date
df_add = df_add[df_add['DateTrans'] > last_cutoff_date]

# BUT NOW AFTER HERE OUR ACTUAL LAST DATE IS the last in df_add
last_cutoff_date = df_add['DateTrans'].max() # Update for df_add !!!!

# Print updates
print('\n|||>_updt.py_> > > Processing Portfolio UPDATES for up to :\n')
print(f"Date Processed: {last_cutoff_date .strftime('%Y-%m-%d')}, Rows processed: {len(df_add)}\n\n")

last_cutoff_date_str = 'AfterAllCEA' # NEEDS to always have the same name 

# Placeholder for OPENING CASH EXCESS UPDTS
filter_cash_excess_check = 0  # MEANS OPEN LOOP FOR UPDTS CASH EXCESS
# #Correct the use of exec function
with open("updt.py", "r", encoding="utf-8") as file:
    code = file.read()
    exec(code)
#print(df)
input('>>>>>>>>>>>>')

```
