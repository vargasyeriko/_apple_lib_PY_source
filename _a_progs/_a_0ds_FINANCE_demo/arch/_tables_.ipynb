```python
import yfinance as yf
import pandas as pd

def fetch_data_with_dividends(symbols, start_date, end_date):
    # Exclude 'CASH' from symbols as it's no longer relevant
    symbols_without_cash = [symbol for symbol in symbols if symbol != 'CASH']
    
    # Initialize a DataFrame to store all data
    all_data = pd.DataFrame()
    
    for symbol in symbols_without_cash:
        # Download historical price data for the symbol
        ticker = yf.Ticker(symbol)
        hist = ticker.history(start=start_date, end=end_date)
        
        # Extracting open, close, and dividends
        data = hist[['Open', 'Close', 'Dividends']]
        
        # Creating a MultiIndex for columns with the symbol name as the top level
        data.columns = pd.MultiIndex.from_product([[symbol], data.columns])
        
        # If all_data is empty, initialize it with data
        if all_data.empty:
            all_data = data
        else:
            # Otherwise, join this symbol's data with the existing all_data
            all_data = all_data.join(data, how='outer')
    
    # Remove days without any trading data across all symbols
    all_data.dropna(how='all', subset=[idx for idx in all_data.columns if 'Open' in idx or 'Close' in idx], inplace=True)
    
    return all_data

# Example usage
start_date = '2022-07-01'  # Start date in 'YYYY-MM-DD' format
end_date = '2024-04-01'  # End date in 'YYYY-MM-DD' format

# Call the function to get price and dividends data
all_data_open_close_dividends = fetch_data_with_dividends(symbols, start_date, end_date)
all_data_open_close_dividends
```

```python
import pandas as pd

def dividend_count(symbols, start_date, end_date):
    # Initialize a dictionary to store dividend counts
    dividend_counts = {}

    for symbol in symbols:
        # Fetch the historical data for the symbol
        ticker = yf.Ticker(symbol)
        hist = ticker.history(start=start_date, end=end_date)
        
        # Count non-zero dividends and add to the dictionary
        non_zero_dividends = hist['Dividends'].astype(bool).sum()
        dividend_counts[symbol] = non_zero_dividends
    
    # Convert the dictionary to a DataFrame for better readability
    dividend_counts_df = pd.DataFrame(list(dividend_counts.items()), columns=['Symbol', 'Dividend Count'])
    
    return dividend_counts_df
```

```python
dividend_counts_data = dividend_count(symbols, start_date, end_date)
dividend_counts_data
```
