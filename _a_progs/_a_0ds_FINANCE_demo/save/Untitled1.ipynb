```python
import matplotlib.pyplot as plt

# Variables
cash = 34
div = 50

# Data to plot
labels = ['Cash', 'Dividends']
sizes = [cash, div]
colors = ['limegreen', 'peachpuff']  # Custom colors
explode = (0.1, 0)  # Explode the 1st slice (Cash)
autopct = '%1.1f%%'  # Percentage format
shadow = True
startangle = 140
textprops = {'fontsize': 12}  # Font size of the labels

# Figure and axis dimensions
figsize_x = 3
figsize_y = 3
fig, ax = plt.subplots(figsize=(figsize_x, figsize_y))

# Pie chart
ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct=autopct,
       shadow=shadow, startangle=startangle, textprops=textprops)

# Aspect ratio
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Values for dividends and cash
dividends = 1200  # Example value for dividends
cash = 800        # Example value for cash

# Labels and values for the pie chart
labels = ['Dividends', 'Cash']
sizes = [dividends, cash]
colors = ['#ff9999','#66b3ff']  # Example colors for each slice

# Figure name
figure_name = "portfolio_distribution"

# Creating the pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.title('Portfolio Distribution: Dividends vs Cash')

# Adding a table at the bottom of the axes
col_labels = ['Dividends', 'Cash']
row_labels = ['Amount']
cell_text = [sizes]
table = ax.table(cellText=cell_text, colLabels=col_labels,
                 rowLabels=row_labels,
                 loc='bottom', cellLoc='center', rowLoc='center',
                 bbox=[0.25, -0.5, 0.5, 0.3])  # Adjust the bbox to fit your layout

# Adjusting subplots to fit the table
plt.subplots_adjust(left=0.1, bottom=0.2, right=0.9, top=0.8)

# Save and show the figure
plt.savefig(f"{figure_name}.png")  # Save the figure with the figure name
plt.show()
```
